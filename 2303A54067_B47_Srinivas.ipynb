```python
# 1) Data Uploading
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/mall customer.csv')
df1 = data
```

```python
categorical_columns = df1.select_dtypes(include=['object']).columns
unique_vals = {col: df1[col].unique() for col in categorical_columns}
```

```python
# 2) Data Visualization
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Scatter Plot: Annual Income vs Spending Score')
plt.show()
```

```python
# 3) Data Analysis
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_features = ['Annual Income (k$)', 'Spending Score (1-100)']
df1_scaled = scaler.fit_transform(df1[numerical_features])
```

```python
# 4. Handling Missing Values
df1 = df1.dropna()
```

```python
# 5. Model Training and Evaluation
from sklearn.cluster import KMeans
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df1_scaled)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
